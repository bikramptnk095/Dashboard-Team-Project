```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
writer = pd.ExcelWriter('Aurostats.xlsx', engine = 'xlsxwriter')
```

```python
intervention= pd.read_csv('intervention.csv', header = None, sep=',')
intervention.columns = ['Id','category_id','description','incident_time','latitude','longitude']
intervention.incident_time= pd.to_datetime(intervention.incident_time)
temp = pd.DatetimeIndex(intervention.incident_time)
intervention['incident_date'] = temp.date
intervention['incident_time'] = temp.time
intervention.isnull().sum()

top_left = [37.351128,-121.951904]
top_right= [37.351128, -121.934616]
bottom_left =[37.343296, -121.951904]
bottom_right = [37.343296, -121.934616]


n = 25

length = top_right[1] - top_left[1]
width = top_left[0] - bottom_left[0]

len_array = [top_left[1]]
wid_array = [bottom_left[0]]

group = []
for i in range(1, n + 1):
    len_array.append(len_array[i - 1] + length/n)
    wid_array.append(wid_array[i - 1] + length/n)


for x in range(0, intervention.shape[0]):
    a = b = 0
    if(intervention.longitude[x] != 0):
        for i in range(1, n + 1):
            a += 1
            for j in range(1, n + 1):
                b += 1
                if(intervention.longitude[x] >= len_array[i - 1] and intervention.longitude[x] <= len_array[i] and intervention.latitude[x] >= wid_array[j - 1] and intervention.latitude[x] <= wid_array[j]):
                    group.append(a + b - 1)
                   
    else:
        group.append(None)

intervention['Location Group'] = group
intervention.to_excel(writer, sheet_name = 'Intervention', index = False)
intervention.head()
```

```python
ride = pd.read_csv('ride.csv',header= None)

ride.columns = ['id','seat_id','time_board','time_offboard','onboard_lat','onboard_lon','offboard_lat','offboard_lon','star_rating','rider_first_name','rider_last_name','rider_email','survey_ride_reason','survey_feedback']
ride = ride[['id','time_board','time_offboard','onboard_lat','onboard_lon','offboard_lat','offboard_lon','star_rating','rider_first_name','rider_last_name']]
ride['time_board'] = pd.to_datetime(ride['time_board'])
ride['time_offboard'] = pd.to_datetime(ride['time_offboard'])
ride['date_board'] = ride['time_board'].dt.date
ride['time_board'] = ride['time_board'].dt.time
ride['date_offboard'] = ride['time_offboard'].dt.date
ride['time_offboard'] = ride['time_offboard'].dt.time
ride = ride.dropna(subset=['onboard_lat'], how='any')
ride = ride.dropna(subset=['onboard_lon'], how='any')
ride = ride.dropna(subset=['offboard_lat'], how='any')
ride = ride.dropna(subset=['offboard_lon'], how='any')
ride.to_excel(writer, sheet_name = 'Ride', index = False)
ride.head()
```
